```python
class Producto:
    def __init__(self, nombre, precio_inicial, stock_inicial):
        self.nombre = nombre
        self._precio = 0
        self.precio = precio_inicial
        self._stock = 0
        self.stock = stock_inicial

    # Precio
    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, nuevo_precio):
        if not isinstance(nuevo_precio, (int, float)):
            raise TypeError("El precio debe ser un n√∫mero")
        if nuevo_precio < 0:
            raise ValueError("El precio no puede ser negativo")
        else:
            self._precio = nuevo_precio

    # Stock
    @property
    def stock(self):
        return self._stock

    @stock.setter
    def stock(self, nuevo_stock):
        if not isinstance(nuevo_stock, int):
            raise TypeError("El stock debe ser un n√∫mero entero")
        if nuevo_stock < 0:
            raise ValueError("El stock no puede ser negativo")
        else:
            self._stock = nuevo_stock

    def vender(self, cantidad):
        if not isinstance(cantidad, int):
            raise TypeError("La cantidad vendida debe ser un n√∫mero entero")
        if cantidad <= 0:
            raise ValueError("La cantidad vendida debe ser mayor que cero")
        if cantidad > self._stock:
            raise ValueError(f"No hay suficiente stock de {self.nombre} (stock actual: {self._stock})")
        else:
            self._stock -= cantidad
            print(f"Se han vendido {cantidad} unidades de {self.nombre}. Stock actual: {self._stock}")
```

```python
from abc import ABC, abstractmethod
"""Clases abstractas e interfaces (abc module)"""

class MedioDePago (ABC):
      def __init__(self, monto):
         self.monto = monto
         print (f"DEBUG: MedioDePago base para monto {self.monto} creado.")

      @abstractmethod

      def procesar_pago(self):
         """para procesar pagos """
      pass



      @abstractmethod

      def obtener_tipo(self):
        """ devuelve el tipo de pago"""
      pass

class TarjetaCredito(MedioDePago):

      def __init__(self, monto, numero_tarjeta, fecha_vencimiento):
        super().__init__(monto)
        self.numero_tarjeta = numero_tarjeta
        self.fecha_vencimiento = fecha_vencimiento
        print(f"[DEBUG]: TarjetaCredito para monto {self.monto} creada.")

      def procesar_pago (self):
        ultimos_4_digitos = self.numero_tarjeta [-4:]
        print(f"Procesando pago de ${self.monto:.2f} con Tarjeta de Cr√©dito (XXXX-XXXX-XXXX-{ultimos_4_digitos}) vencimiento {self.fecha_vencimiento}.")


      def obtener_tipo(self):
        return "Tarjeta de Cr√©dito"


class Efectivo(MedioDePago):
      def __init__(self, monto, recibido):
        super().__init__(monto)
        self.recibido = recibido
        print(f"DEBUG: Pago en Efectivo por monto {self.monto} y recibido {self.recibido}.")


      def procesar_pago(self):

          cambio = self.recibido - self.monto
          print(f"Pago de ${self.monto:.2f} en Efectivo. Se recibi√≥ ${self.recibido:.2f}. Cambio: ${cambio:.2f}.")
          if cambio < 0 :
            print("Falta dinero para completar el pago.")

      def obtener_tipo(self):
        return "Efectivo"







```

```python
""" Decoradores y su aplicaci√≥n en POO """


"""Ejercicio: Decorador para Logging Autom√°tico üìù
Objetivo: Crear un decorador que autom√°ticamente registre (imprima en consola) informaci√≥n sobre una funci√≥n cada vez que es llamada: su nombre, los argumentos con los que fue llamada, y el valor que retorna.

Este es un caso de uso muy com√∫n en el desarrollo de software para depuraci√≥n y monitoreo.

Requisitos:

Define el Decorador log_llamada:

Crea una funci√≥n llamada log_llamada que reciba como argumento la funci√≥n (func) que va a decorar.

Dentro de log_llamada, define una funci√≥n anidada llamada wrapper que acepte *args y **kwargs para ser flexible con los argumentos de la funci√≥n original.

L√≥gica del wrapper:

Antes de ejecutar func: Imprime un mensaje que indique que la funci√≥n est√° siendo llamada, incluyendo su nombre y los argumentos que recibi√≥.

Formato sugerido: "[LOG]: Llamando a '{nombre_funcion}' con args={args}, kwargs={kwargs}"

Pista: Puedes obtener el nombre de la funci√≥n con func.__name__.

Ejecuta func: Llama a la funci√≥n original (func) con los *args y **kwargs recibidos y guarda su resultado.

Despu√©s de ejecutar func: Imprime un mensaje que indique que la funci√≥n ha finalizado y muestre el resultado que devolvi√≥.

Formato sugerido: "[LOG]: '{nombre_funcion}' finaliz√≥. Resultado: {resultado}"

Retorna el resultado: El wrapper debe devolver el resultado de la func original.

Aplica el Decorador:

Usa el decorador @log_llamada sobre al menos dos funciones diferentes que crees. Por ejemplo:

Una funci√≥n simple que sume dos n√∫meros.

Una funci√≥n que simule una operaci√≥n m√°s compleja (ej. un factorial o una b√∫squeda).

Llama a estas funciones para ver el decorador en acci√≥n."""


def log_llamada(func):
    def wrapper(*args, **kwargs):
        print(f"Llamada a la funci√≥n: {func.__name__} con args={args}, kwargs={kwargs}")
        resultado = func(*args, **kwargs)
        print(f"La funcion {func.__name__} ha finalizado y este es el resultado: {resultado}")

        return resultado

    return wrapper

@log_llamada

def suma(a,b):
    print (f" realizando suma de {a} + {b}")
    return a + b


@log_llamada

def saludar (nombre, saludo="HOLA"):
  print (f" generando saLUDO PARA {nombre}...")
  return f"{saludo}, {nombre}"





# --- Llama a tus funciones para probar el decorador ---

print("--- Prueba 1: Suma ---")
resultado_suma = suma(5, 3)
print(f"El resultado final de la suma es: {resultado_suma}\n")

print("--- Prueba 2: Saludo ---")
mensaje_saludo = saludar("Alice", saludo="¬°Qu√© tal")
print(f"El mensaje de saludo final es: {mensaje_saludo}\n")

print("--- Prueba 3: Saludo con valor por defecto ---")
mensaje_saludo_default = saludar("Bob")
print(f"El mensaje de saludo final es: {mensaje_saludo_default}\n")
```
